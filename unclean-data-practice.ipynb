## Checking and cleaning data

### Incosistent Column Names

```python
# First of all i fix the index number to start from 1 and not from 0
df.index = np.arange(1, len(df) + 1)
```

```python
df
```

```python
# Secondly, i remove the annoying symbols from movies { }
df['movie_title'] = df['movie_title'].str.replace('?', '')
df['movie_title'] = df['movie_title'].str.replace('Ã¿', '')
```

```python
df
```

```python
# We need to change cases (upper or lower) and then rename
df.columns
```

```python
# we convert all letters in columns description in upper
df.columns = df.columns.str.upper()
df.columns
```

```python
df
```

```python
# rename a column {for example the DURATION to TIME}
df.rename(columns= {'DURATION':'TIME'})
```

### Missing Data

```python
# first we check for missing data

df.isnull().sum()
```

```python
x = round(df['DURATION'].mean())
```

```python
# Replace NaN values in DURATION collumn with average duration value

df1 = df.fillna({"DURATION": x})
df1
```

```python
df1.isnull().sum()
```

```python
df
```

```python
# FACENUMBER_IN_POSTER and TITLE_YEAR collumns have many NaN values ( 5 and 7) so i choose to delete these 2 collumns

del df1["FACENUMBER_IN_POSTER"]
del df1["TITLE_YEAR.1"]
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
# For the left remaining NaN values i choose to fill them with fillna method
# and forward fill method

df2 = df1.fillna(method="ffill")
```

```python
df2.isnull().sum()
```

```python
df2
```

### Dropping Duplicates

```python
# In movie collumn i found duplicated rows

df3 = df2.drop_duplicates('MOVIE_TITLE')
```
