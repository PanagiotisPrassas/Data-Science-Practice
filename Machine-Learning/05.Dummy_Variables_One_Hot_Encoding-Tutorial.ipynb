```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sn
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/dummy-variables-and-one-hot-enconding-csv/homeprices(2).csv')
```

```python
df
```

# Create dummies variables

```python
dummies = pd.get_dummies(df['town'], dtype=int)
```

```python
dummies
```

```python
# We merge the df with the dummies df

merged = pd.concat([df, dummies], axis='columns')
```

```python
merged
```

```python
# We drop now, the town collumn and we drop one collumn of dummies so we are not fall in dummy variable trap!

final = merged.drop(['town','west windsor'], axis = 'columns')
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x = final.drop('price', axis = 'columns')
```

```python
x
```

```python
y = final['price']
```

```python
y
```

```python
model.fit(x,y)
```

```python
# 2800 sqr ft home in robbinsville

model.predict([[2800,0,1]])
```

```python
# 3400 sqr ft home in west windsor

model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```

# Using sklearn OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dfle =  df
dfle['town'] = le.fit_transform(dfle['town'])
```

```python
dfle
```

```python
x2 = dfle[['town', 'area']].values
```

```python
x2
```

```python
y2 = dfle['price'].values
```

```python
y2
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x2 = ct.fit_transform(x2)
```

```python
x2
```

```python
x2 = x2[:,1:]
```

```python
model.fit(x2,y2)
```

```python
# 3400 sqr ft home in west windsor

model.predict([[0,1,3400]])
```

```python
# 2800 sqr ft home in robbinsville

model.predict([[1,0,2800]])
```
