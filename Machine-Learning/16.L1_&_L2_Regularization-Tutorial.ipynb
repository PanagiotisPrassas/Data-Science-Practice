```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/kaggle/input/regularization-l1-and-l2-tutorial/Melbourne_housing_FULL.csv')
```

```python
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
# let's use limited columns which makes more sense for serving our purpose

columns_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[columns_to_use]
```

```python
df.head()
```

```python
# check for nan values

df.isna().sum()
```

```python
# fill zeros in some columns

columns_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[columns_to_fill_zero] = df[columns_to_fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
# in some columns we will fill with mean value

df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
df = pd.get_dummies(df, dtype='int', drop_first=True)
df.head()
```

```python
X = df.drop('Price', axis='columns')
y = df.Price
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

# Using Lasso (L1 Regularized) Regression Model

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

# Using Ridge (L2 Regularized) Regression Model

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```
