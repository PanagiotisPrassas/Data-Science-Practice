**Exercise - Predict Income in year 2023**

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sn
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/canada-per-capita-income/canada_per_capita_income.csv')
```

```python
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'], color='red')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2023]])
```

**The predicted income for year 2023 is 43774.08$ !**
