**Exercise - In exercise folder (same level as this notebook on github) there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates**

**2 yr experience, 9 test score, 6 interview score**

**12 yr experience, 10 test score, 10 interview score**

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sn
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/multivariable-exercise/hiring.csv')
```

```python
df
```

```python
# At experience collumn by filling nan values , for purpose reasons we fill them with zero

df['experience'] = df['experience'].fillna('zero')
```

```python
df
```

```python
# Convert the strings to numbers in experience collumn
# We need w2n function

!pip install word2number
from word2number import w2n
```

```python
df['experience'] = df['experience'].apply(w2n.word_to_num)
```

```python
df
```

```python
# Filling nan values in test_score collumn with median value

df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
```

```python
df['test_score(out of 10)'].mean()
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Find salary with 2 yr experience, 9 test score, 6 interview score

reg.predict([[2, 9, 6]])
```

**The predicted salary is 53205($) !**

```python
# Find salary with 12 yr experience, 10 test score, 10 interview score

reg.predict([[12, 10, 10]])
```

**The predicted salary is 92002($) !**
