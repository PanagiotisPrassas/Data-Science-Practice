**1) Predict price of a mercedez benz that is 4 yr old with mileage 45000**

**2) Predict price of a BMW X5 that is 7 yr old with mileage 86000**

# Import libraries and Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sn
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/dummy-variables-and-exercise/carprices.csv')
```

```python
df
```

# Data Visualization

```python
# Plot a scatterplot

plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.scatter(df['Mileage'],df['Sell Price($)'], color='red')
plt.show()
```

# Create dummies variables

```python
dummies = pd.get_dummies(df['Car Model'], dtype = int)
```

```python
dummies
```

```python
# We merge the df with the dummies df

merged = pd.concat([df, dummies], axis='columns')
```

```python
merged
```

```python
# We drop now, the town collumn and we drop one collumn of dummies so we are not fall in dummy variable trap!

final = merged.drop(['Car Model','Mercedez Benz C class'], axis = 'columns')
```

```python
final
```

# Proceed with model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x = final.drop('Sell Price($)', axis = 'columns')
```

```python
x
```

```python
y = final['Sell Price($)']
```

```python
y
```

```python
model.fit(x,y)
```

```python
# Predict price of a mercedez benz that is 4 yr old with mileage 45000

model.predict([[45000,4,0,0]])
```

**The predicted price is 36991$ !**

```python
# Predict price of a BMW X5 that is 7 yr old with mileage 86000

model.predict([[86000,7,0,1]])
```

**The predicted price is 11080$ !**
