# Step 2: Remove outliers

```python
df1 = df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())]
df1 = df[df.RestingBP<(df.RestingBP.mean()+3*df.RestingBP.std())]
```

```python
df2 = df1[df1.Cholesterol>(df1.Cholesterol.mean()+3*df1.Cholesterol.std())]
df2 = df1[df1.Cholesterol<(df1.Cholesterol.mean()+3*df1.Cholesterol.std())]
```

```python
df3 = df2[df2.FastingBS>(df2.FastingBS.mean()+3*df2.FastingBS.std())]
df3 = df2[df2.FastingBS<(df2.FastingBS.mean()+3*df2.FastingBS.std())]
```

```python
df4 = df3[df.MaxHR>(df3.MaxHR.mean()+3*df3.MaxHR.std())]
df4 = df3[df.MaxHR<(df3.MaxHR.mean()+3*df3.MaxHR.std())]
```

```python
df5 = df4[df.Oldpeak>(df4.Oldpeak.mean()+3*df4.Oldpeak.std())]
df5 = df4[df.Oldpeak<(df4.Oldpeak.mean()+3*df4.Oldpeak.std())]
```

# Step 3: Convert text columns to numbers

```python
df5.head()
```

```python
df5.Sex.unique()
```

```python
df5.ChestPainType.unique()
```

```python
df5.RestingECG.unique()
```

```python
df5.ExerciseAngina.unique()
```

```python
df5.ST_Slope.unique()
```

```python
df6 = df5.copy()

df6.ChestPainType.replace(
    {
        'ATA': 0,
        'NAP': 1,
        'ASY': 2,
        'TA' : 3
    },
    inplace=True
)

df6.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True
)

df6.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True
)

df6.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df6.head()
```

```python
df7 = pd.get_dummies(df6, dtype='int', drop_first=True)
df7.head()
```

# Step 4: Apply Scaling

```python
X = df7.drop('HeartDisease', axis='columns')
y = df7.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

# Step 5: Build a classification model using support vector machine. Use standalone model as well as Bagging model and check if you see any difference in the performance.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, random_state=10)
```

```python
from sklearn import svm
from sklearn.model_selection import cross_val_score

scores = cross_val_score(svm.SVC(), X, y, cv=5)
scores
```

```python
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator=svm.SVC(), 
    n_estimators=100, 
    max_samples=0.8, 
    oob_score=True,
    random_state=0
)
scores = cross_val_score(bag_model, X, y, cv=5)
scores
```
