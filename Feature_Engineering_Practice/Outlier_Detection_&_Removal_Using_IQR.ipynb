# Step 2: Using IQR detect weight outliers and print them

```python
# Calculate Q1 and Q3 

Q1 = df['Weight'].quantile(0.25)
Q3 = df['Weight'].quantile(0.75)
```

```python
Q1, Q3
```

```python
# Calculate IQR

IQR = Q3 - Q1
```

```python
IQR
```

```python
min_threshold = Q1 - 1.5*IQR
max_threshold = Q3 + 1.5*IQR
```

```python
min_threshold, max_threshold
```

```python
# So the outliers are: 

df[(df['Weight'] < min_threshold) | (df['Weight'] > max_threshold)]
```
