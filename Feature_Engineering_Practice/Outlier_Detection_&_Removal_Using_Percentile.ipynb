# Exercise on practicing with removing outliers using percentile based on price per night for a given apartment/home

```python
# import libraries

import numpy as np
import pandas as pd
```

```python
df =  pd.read_csv('/kaggle/input/new-york-city-airbnb-open-data/AB_NYC_2019.csv')
```

```python
df.shape
```

```python
df['price'].describe()
```

```python
df['price'].quantile(0.0001)
```

```python
df['price'].quantile(0.99)
```

```python
# We need to remove zero from prices and the high prices from the dataset

min_percentile = df['price'].quantile(0.0001)
max_percentile = df['price'].quantile(0.99)
```

```python
new_df = df[df['price'] > min_percentile]
new_df = df[df['price'] < max_percentile]
```

```python
new_df.shape
```

```python
new_df['price'].describe()
```

**So, by comparing the shapes of old and new dataframe we found ->  48895 - 48403 = 492 outliers !**
